```python
import os, glob
import pandas as pd

folder = 'Guidelines'
abdico_files = '*_ABDICO.csv'

# load and concat files

dfs = []
files = []
files.extend(glob.glob(os.path.join(folder, abdico_files)))

columns = ["input_statement","statement_type","A","D","I","B","C","O"]
for f in files:
    df = pd.read_csv(f)
    # Add a column with the filename stripped from '_sentences_ABDICO.csv'
    base_name = os.path.basename(f)
    stripped_name = base_name.replace('_sentences_ABDICO.csv', '')
    df['source_file'] = stripped_name
    dfs.append(df[['source_file']+ columns].copy())
all_df = pd.concat(dfs, ignore_index=True)
```

```python
def column_frequency_table(df, column_name):
    """Returns a frequency table (DataFrame) for the given column in df."""
    freq = df[column_name].value_counts().reset_index()
    freq.columns = [column_name, 'frequency']
    return freq

statement_type_freq = column_frequency_table(all_df, 'statement_type')
print(statement_type_freq)
```

```python
def count_nonempty_rows(df, columns):
    """Returns the count of rows where all specified columns are not NaN or empty."""
    # For each row, check that all columns are not NaN and not empty string
    mask = df[columns].notna() & (df[columns] != '')
    # Only count rows where all columns are nonempty
    count = mask.all(axis=1).sum()
    return count

# Example usage for columns ['A', 'D', 'I', 'C']:
adic_count = count_nonempty_rows(all_df, ['A', 'D', 'I', 'C'])
print(f"Rows with nonempty A, D, I, C: {adic_count}")
```

```python
def get_nonempty_rows(df, columns):
    """Returns the DataFrame of rows where all specified columns are not NaN or empty."""
    mask = df[columns].notna() & (df[columns] != '')
    sliced_df = df[mask.all(axis=1)]
    return sliced_df

# Example usage for columns ['A', 'D', 'I', 'C']:
adic_rows = get_nonempty_rows(all_df, ['A', 'D', 'I', 'C'])
# adic_rows
# save in csv 
adic_rows.to_csv('adic_nonempty_rows.csv', index=False)
adic_rows
```
